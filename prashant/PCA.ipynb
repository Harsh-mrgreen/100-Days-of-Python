```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import pandas as pd
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df
```

```python
X = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
```

```python
X
```

```python
X.mean()
```

```python
X_centered = X-X.mean()
```

```python
cov_X = np.dot(X_centered.T,(X_centered))
```

```python
cov_X
```

```python
(eig_val, eig_vec) = np.linalg.eig(cov_X)
```

```python
eig_val
```

```python
eig_vec
```

```python
eig_val/eig_val.sum()
```

```python
eig_vec
```

```python
EigVec_2 = eig_vec[:,[0,1]]
```

```python
np.dot(X_centered,EigVec_2)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
transformed_data = pca.fit(X).transform(X)
```

```python
transformed_data
```

```python
# Explained Variance
pca.explained_variance_
```

```python
pca.explained_variance_/pca.explained_variance_.sum()
```

```python
# Factor Loadings: Correlation between each variable and the principle component:
pca.components_.T * np.sqrt(pca.explained_variance_)
```

```python
# Communality: % Variability that the selected PCA's would explain each variable:
Factor_Loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
```

```python
Factor_Loadings[:,0]**2+Factor_Loadings[:,1]**2
```
